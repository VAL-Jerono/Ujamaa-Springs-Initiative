Macro and Weighted Average

The macro average F1-score is 0.42, reflecting the unweighted average of F1-scores for all classes. The weighted average F1-score is 0.60, considering the number of instances for each class. The weighted average is slightly higher due to the imbalanced nature of the classes.

The model performs reasonably well in predicting functional and non-functional wells but struggles with wells needing repair, likely due to the imbalanced dataset.
Further optimization and exploration, such as addressing class imbalance or trying more complex models, may improve the model's ability to predict wells needing repair.
